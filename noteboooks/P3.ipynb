# Algorytmy generowania i weryfikacji danych losowych

---

## Wprowadzenie

Dane losowe są wykorzystywane w wielu obszarach informatyki, między innymi w symulacjach, testowaniu oprogramowania, analizie danych oraz kryptografii. Ich jakość ma duże znaczenie, ponieważ wpływa na poprawność działania algorytmów, wiarygodność wyników oraz poziom bezpieczeństwa systemów.

Celem projektu jest omówienie wybranych algorytmów generowania danych losowych oraz sposobów ich weryfikacji. W ramach obecnego etapu skupiono się na opisie algorytmów planowanych do wykorzystania, przygotowaniu wstępnych danych testowych oraz implementacji pierwszych elementów środowiska badawczego w Jupyter Notebook z użyciem gotowych bibliotek.

Projekt ma stanowić podstawę do dalszej analizy jakości generowanych danych oraz porównania różnych podejść do ich tworzenia i oceny.

## Wybór algorytmów do analizy i implementacji

W ramach obecnego etapu projektu wybrano ograniczony zestaw algorytmów, który pozwala porównać różne podejścia do generowania danych losowych, a jednocześnie pozostaje możliwy do wygodnej implementacji i analizy w środowisku Jupyter Notebook. Zamiast obejmować od razu wszystkie rozważane klasy generatorów, zdecydowano się na wybór kilku reprezentatywnych rozwiązań z grupy PRNG oraz CSPRNG. Takie podejście pozwala skupić się na jakości generowanych danych, różnicach między algorytmami oraz przygotowaniu spójnej struktury programu do dalszej rozbudowy.

W pierwszej kolejności do implementacji i porównania wybrano następujące algorytmy:
- LCG (Linear Congruential Generator),
- Mersenne Twister (MT19937),
- PCG (Permuted Congruential Generator),
- generator systemowy oparty na `os.urandom` lub module `secrets` jako przykład generatora kryptograficznie bezpiecznego.


Na aktualnym etapie rozwazana jest równiez pseudo-implementacja generatroa typu TRNG poprzez zdefiniowanie funkcji mockującej fizyczne źródło entropii oraz przygotowanie generatora opartego o mapy chaotyczne. Szczegóły dotyczące ich implementacji zostaną doprecyzowane w dalszych etapach. 

### 1. LCG (Linear Congruential Generator)

LCG jest jednym z najprostszych i najstarszych generatorów pseudolosowych. Jego działanie opiera się na rekurencyjnym wzorze:

`X_(n+1) = (aX_n + c) mod m`

gdzie:
- `X_n` oznacza aktualny stan generatora,
- `a` jest mnożnikiem,
- `c` jest przyrostem,
- `m` jest modułem.

Zaletą tego generatora jest bardzo prosta konstrukcja, niewielkie wymagania obliczeniowe oraz łatwość zrozumienia działania.

Jednocześnie LCG ma istotne ograniczenia. Przy źle dobranych parametrach może generować sekwencje o słabych własnościach statystycznych, krótkim okresie i widocznych zależnościach między kolejnymi wartościami. Z tego powodu nie jest odpowiedni do zastosowań kryptograficznych ani do bardziej wymagających analiz. Został jednak wybrany celowo, ponieważ stanowi dobry przykład prostego generatora, którego zachowanie można łatwo porównać z nowocześniejszymi rozwiązaniami. 

### 2. Mersenne Twister (MT19937)

Mersenne Twister to jeden z najbardziej znanych generatorów pseudolosowych ogólnego przeznaczenia. Charakteryzuje się bardzo długim okresem, wysoką wydajnością oraz dobrymi własnościami statystycznymi w wielu typowych zastosowaniach. Jest szeroko wykorzystywany w symulacjach, analizie numerycznej i środowiskach programistycznych; w szkicu projektu został wskazany między innymi jako generator używany w module `random` języka Python. 

Generator ten został wybrany, ponieważ stanowi naturalny krok dalej względem prostego LCG. Pozwala pokazać, jak nowocześniejsze PRNG poprawiają jakość generowanych sekwencji i eliminują część problemów charakterystycznych dla starszych konstrukcji. 

Wadą Mersenne Twistera jest brak bezpieczeństwa kryptograficznego. Mimo dobrych własności statystycznych generator ten nie został zaprojektowany z myślą o odporności na przewidywanie i odtwarzanie stanu. Z tego względu dobrze nadaje się do porównania z generatorem kryptograficznie bezpiecznym, który zostanie uwzględniony w projekcie jako osobna klasa rozwiązania.

### 3. PCG (Permuted Congruential Generator)

PCG jest nowocześniejszym generatorem pseudolosowym, który łączy prostotę konstrukcji z lepszą jakością generowanych sekwencji niż klasyczny LCG. W dotychczasowym materiale projektowym został wskazany jako bardziej nowoczesna alternatywa dla starszych generatorów oraz rozwiązanie o niewielkim rozmiarze stanu.

Z punktu widzenia projektu PCG jest cennym wyborem, ponieważ pozwala rozszerzyć porównanie o generator reprezentujący współczesne podejście do PRNG. Dzięki temu możliwe będzie zestawienie:
- bardzo prostego generatora klasycznego,
- popularnego generatora ogólnego przeznaczenia,
- nowszego generatora zoptymalizowanego pod względem jakości i prostoty.


### 4. Generator systemowy `os.urandom` / `secrets`

Jako przedstawiciela generatorów kryptograficznie bezpiecznych wybrano generator systemowy dostępny z poziomu Pythona przez `os.urandom` lub moduł `secrets`. W dotychczasowym szkicu projektu jako przykłady generatorów CSPRNG wskazano między innymi rozwiązania takie jak Fortuna, ChaCha20 oraz HMAC-DRBG, a w proponowanej strukturze programu pojawił się także wrapper na `os.urandom`.

Wybór generatora systemowego ma charakter praktyczny. Pozwala uwzględnić w badaniu klasę CSPRNG bez konieczności samodzielnej implementacji mechanizmów kryptograficznych. Taki generator jest odpowiedni jako punkt odniesienia dla bezpieczeństwa i jakości losowości w zastosowaniach, gdzie zwykły PRNG nie jest wystarczający.

## Uzasadnienie końcowego wyboru

Ostateczny zestaw algorytmów został dobrany tak, aby objąć dwa najważniejsze podejścia do generowania danych losowych:
- pseudolosowość ogólnego przeznaczenia,
- losowość kryptograficznie bezpieczną.

Wybrane algorytmy tworzą spójny zestaw porównawczy:
- **LCG** jako prosty i historycznie ważny generator referencyjny,
- **MT19937** jako popularny i wydajny generator ogólnego przeznaczenia,
- **PCG** jako nowocześniejszy przedstawiciel PRNG,
- **`os.urandom` / `secrets`** jako praktyczny przykład CSPRNG.

Taki dobór pozwala pokazać różnice między prostotą implementacji, jakością statystyczną oraz bezpieczeństwem generatorów. Jednocześnie zakres ten jest wystarczająco mały, aby możliwe było przygotowanie pierwszej działającej wersji projektu, zgromadzenie danych testowych i przeprowadzenie wstępnej weryfikacji wyników. Szersze klasy generatorów, takie jak DRBG zgodne z NIST, generatory chaotyczne czy TRNG, mogą zostana doprecyzowane i uwzględnione na kolejnych etapach projektu.
