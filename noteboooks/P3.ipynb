# Algorytmy generowania i weryfikacji danych losowych

---

## Wprowadzenie

Dane losowe są wykorzystywane w wielu obszarach informatyki, między innymi w symulacjach, testowaniu oprogramowania, analizie danych oraz kryptografii. Ich jakość ma duże znaczenie, ponieważ wpływa na poprawność działania algorytmów, wiarygodność wyników oraz poziom bezpieczeństwa systemów.

Celem projektu jest omówienie wybranych algorytmów generowania danych losowych oraz sposobów ich weryfikacji. W ramach obecnego etapu skupiono się na opisie algorytmów planowanych do wykorzystania, przygotowaniu wstępnych danych testowych oraz implementacji pierwszych elementów środowiska badawczego w Jupyter Notebook z użyciem gotowych bibliotek.

Projekt ma stanowić podstawę do dalszej analizy jakości generowanych danych oraz porównania różnych podejść do ich tworzenia i oceny.

## Wybór algorytmów do analizy i implementacji

W ramach projektu przyjęto zestaw algorytmów obejmujący najważniejsze klasy generatorów danych losowych, tak aby możliwe było porównanie różnych podejść do uzyskiwania losowości oraz ocena ich przydatności w odmiennych zastosowaniach. Dobór algorytmów został opracowany w taki sposób, aby uwzględnić zarówno klasyczne generatory pseudolosowe, nowocześniejsze konstrukcje o lepszych właściwościach statystycznych, generatory kryptograficznie bezpieczne, jak również rozwiązania inspirowane fizycznymi źródłami entropii oraz dynamiką układów chaotycznych.

Do analizy i implementacji wybrano następujące algorytmy:
- **LCG (Linear Congruential Generator)** jako reprezentanta klasycznych generatorów PRNG,
- **Mersenne Twister (MT19937)** jako popularny generator pseudolosowy ogólnego przeznaczenia,
- **PCG (Permuted Congruential Generator)** jako przykład nowocześniejszego generatora PRNG,
- **generator systemowy oparty na `os.urandom` lub module `secrets`** jako przedstawiciela generatorów kryptograficznie bezpiecznych,
- **generator inspirowany TRNG**, oparty na symulowanym źródle fizycznej entropii,
- **generator oparty na mapie chaotycznej** jako reprezentanta generatorów bazujących na zjawiskach nieliniowych.

Taki zestaw pozwala objąć każdą z głównych kategorii generatorów rozważanych w projekcie i stworzyć spójną podstawę do dalszych testów jakości losowości, analizy statystycznej oraz porównań praktycznych.

### 1. LCG (Linear Congruential Generator)

LCG jest jednym z najprostszych i najstarszych generatorów pseudolosowych. Jego działanie opiera się na rekurencyjnym wzorze:

`X_(n+1) = (aX_n + c) mod m`

gdzie:
- `X_n` oznacza aktualny stan generatora,
- `a` jest mnożnikiem,
- `c` jest przyrostem,
- `m` jest modułem.

Generator ten został wybrany jako reprezentant klasycznych PRNG ze względu na swoją prostotę, przejrzystość działania oraz znaczenie historyczne. Pozwala on łatwo przedstawić podstawowe mechanizmy generowania sekwencji pseudolosowej, rolę ziarna początkowego oraz wpływ parametrów na jakość otrzymywanych wyników.

Jednocześnie LCG ma istotne ograniczenia, takie jak krótszy okres, słabsze właściwości statystyczne oraz podatność na występowanie zależności między kolejnymi wartościami. Dzięki temu stanowi dobry punkt odniesienia w porównaniu z bardziej rozwiniętymi generatorami uwzględnionymi w projekcie.

### 2. Mersenne Twister (MT19937)

Mersenne Twister jest jednym z najbardziej rozpowszechnionych generatorów pseudolosowych ogólnego przeznaczenia. Charakteryzuje się bardzo długim okresem, wysoką wydajnością oraz dobrymi właściwościami statystycznymi w typowych zastosowaniach obliczeniowych. Jest szeroko wykorzystywany między innymi w symulacjach, analizie numerycznej oraz środowiskach programistycznych.

Algorytm ten został uwzględniony w projekcie jako reprezentant dojrzałych generatorów PRNG stosowanych w praktyce. Pozwala on pokazać różnicę między prostymi konstrukcjami, takimi jak LCG, a generatorami zaprojektowanymi z myślą o wysokiej jakości sekwencji i dużym okresie.

Mersenne Twister nie jest jednak generatorem kryptograficznie bezpiecznym, dlatego jego obecność w projekcie ma również znaczenie porównawcze względem generatorów CSPRNG.

### 3. PCG (Permuted Congruential Generator)

PCG jest nowocześniejszym generatorem pseudolosowym, łączącym prostą strukturę z lepszymi właściwościami statystycznymi niż klasyczne rozwiązania, takie jak LCG. Konstrukcja ta została zaprojektowana w taki sposób, aby przy niewielkim stanie wewnętrznym zapewniać dobrą jakość generowanych sekwencji oraz wysoką efektywność działania.

Włączenie PCG do projektu pozwala rozszerzyć analizę o generator reprezentujący współczesne podejście do PRNG. Dzięki temu możliwe staje się porównanie generatora klasycznego, generatora powszechnie używanego oraz nowszego rozwiązania zoptymalizowanego pod względem jakości i prostoty implementacyjnej.

PCG pełni w projekcie rolę pomostu między tradycyjnymi generatorami pseudolosowymi a bardziej zaawansowanymi konstrukcjami stosowanymi obecnie w praktyce.

### 4. Generator systemowy `os.urandom` / `secrets`

Jako przedstawiciela generatorów kryptograficznie bezpiecznych wybrano generator systemowy dostępny z poziomu Pythona przez `os.urandom` lub moduł `secrets`. Rozwiązanie to korzysta z mechanizmów systemu operacyjnego odpowiedzialnych za pozyskiwanie i przetwarzanie entropii, dzięki czemu nadaje się do zastosowań wymagających większej odporności na przewidywanie generowanych wartości.

Uwzględnienie generatora systemowego pozwala objąć w projekcie kategorię CSPRNG bez konieczności samodzielnej implementacji złożonych mechanizmów kryptograficznych. Jest to rozwiązanie praktyczne, łatwo dostępne i jednocześnie istotne z punktu widzenia porównania z generatorami pseudolosowymi ogólnego przeznaczenia.

Jego obecność w zestawie badanych algorytmów umożliwia ocenę różnic między generatorami projektowanymi głównie pod kątem jakości statystycznej a generatorami tworzonymi z myślą o bezpieczeństwie.

### 5. Generator inspirowany TRNG

W projekcie uwzględniono również generator inspirowany sprzętowymi generatorami liczb losowych typu TRNG. Ze względu na ograniczenia środowiska Jupyter Notebook oraz brak bezpośredniego dostępu do fizycznego układu generującego entropię, rozwiązanie to zostanie zrealizowane w formie symulacji źródła fizycznego. Oznacza to przygotowanie funkcji modelującej proces pozyskiwania entropii z niestabilnego zjawiska fizycznego, takiego jak szum, zakłócenia sygnału lub losowe fluktuacje pomiarowe.

Celem uwzględnienia tej kategorii nie jest odtworzenie rzeczywistego układu sprzętowego, lecz przedstawienie idei generatorów prawdziwie losowych oraz pokazanie różnicy między algorytmicznym rozwijaniem sekwencji a pozyskiwaniem entropii ze źródeł zewnętrznych. Takie podejście pozwala zachować spójność metodologiczną projektu, a jednocześnie włączyć do analizy kategorię generatorów opartych o procesy fizyczne.

Generator ten będzie pełnił funkcję demonstracyjną i porównawczą, stanowiąc przybliżenie klasy TRNG możliwe do zaprezentowania w środowisku notebooka.

### 6. Generator oparty na mapie chaotycznej

Ostatnią uwzględnioną kategorią jest generator oparty na mapie chaotycznej. Rozwiązania tego typu wykorzystują właściwości układów nieliniowych, w których bardzo małe zmiany warunków początkowych prowadzą do znacznych różnic w dalszym przebiegu sekwencji. W kontekście generowania danych losowych generatory chaotyczne są interesujące, ponieważ łączą deterministyczny charakter z wysoką wrażliwością na parametry początkowe.

Włączenie generatora chaotycznego do projektu pozwala rozszerzyć analizę o klasę rozwiązań odmienną od klasycznych generatorów PRNG i od generatorów kryptograficznych. Jest to istotne z punktu widzenia porównania różnych mechanizmów prowadzących do uzyskania sekwencji o cechach zbliżonych do losowych.

Generator oparty na mapie chaotycznej będzie w projekcie pełnił rolę rozwiązania eksperymentalnego i badawczego, umożliwiającego ocenę, na ile podejście wykorzystujące dynamikę chaosu może stanowić alternatywę dla bardziej standardowych metod generowania danych losowych.

## Podejście do testowania i weryfikacji generatorów

Testowanie generatorów danych losowych stanowi jeden z kluczowych elementów projektu, ponieważ sama implementacja lub użycie danego algorytmu nie daje jeszcze pewności, że generowane sekwencje mają odpowiednie właściwości statystyczne lub spełniają wymagania konkretnego zastosowania. Z tego względu w projekcie przyjęto podejście wielopoziomowe, obejmujące zarówno testy wstępne realizowane bezpośrednio w środowisku Jupyter Notebook, jak i możliwość wykorzystania bardziej zaawansowanych zewnętrznych narzędzi w dalszych etapach prac.

### 1. Testy wstępne realizowane w notebooku

Na pierwszym etapie przewidziano wykonanie własnych testów podstawowych, które pozwolą szybko porównać działanie wybranych generatorów oraz wykryć najbardziej oczywiste odchylenia od oczekiwanej losowości. Testy te będą realizowane bezpośrednio w Pythonie i obejmą przede wszystkim:
- analizę podstawowych statystyk próby, takich jak średnia, wariancja, minimum i maksimum,
- analizę rozkładu wartości w postaci histogramu,
- test częstości bitów lub wartości,
- test serii,
- prostą ocenę równomierności rozkładu.

Tego rodzaju weryfikacja ma charakter wstępny i diagnostyczny. Jej celem jest szybkie wychwycenie generatorów o wyraźnie słabych właściwościach oraz uzyskanie pierwszego porównania pomiędzy generatorami klasycznymi, kryptograficznie bezpiecznymi, chaotycznymi oraz inspirowanymi źródłami fizycznej entropii.

### 2. Testy zgodne z podejściem NIST SP 800-22

W dalszej części projektu planowane jest uwzględnienie testów opartych na podejściu przedstawionym w NIST SP 800-22. Jest to jeden z najczęściej przywoływanych standardów dotyczących statystycznej oceny generatorów liczb pseudolosowych i losowych. W dokumentacji projektu wskazano przykładowo takie testy jak Longest Run of Ones, Binary Matrix Rank Test oraz Approximate Entropy Test. 

Ze względu na złożoność pełnej baterii testów oraz konieczność odpowiedniego przygotowania danych wejściowych, na obecnym etapie nie zakłada się implementacji całego pakietu od podstaw. Bardziej realistycznym rozwiązaniem jest wykorzystanie wybranych testów lub gotowych implementacji społecznościowych, tak aby móc odwołać się do uznanego standardu i jednocześnie zachować możliwość praktycznego uruchomienia eksperymentów w środowisku projektowym.

### 3. TestU01 jako rozszerzona walidacja empiryczna

Kolejnym poziomem weryfikacji może być wykorzystanie biblioteki TestU01, która została wskazana w materiale projektowym jako jedno z najmocniejszych narzędzi do empirycznego testowania generatorów liczb losowych. Biblioteka ta udostępnia zestawy testów takie jak SmallCrush, Crush oraz BigCrush. 

W projekcie TestU01 może pełnić rolę zaawansowanego narzędzia porównawczego, szczególnie dla generatorów, które pozytywnie przejdą testy wstępne i będą warte dalszej analizy. Z uwagi na większą złożoność integracji tego narzędzia z notebookiem, jego użycie traktowane jest jako etap rozszerzony, służący pogłębionej walidacji jakości sekwencji.

### 4. Dieharder jako drugi poziom walidacji

W dokumentacji projektu uwzględniono również Dieharder, czyli rozwinięcie klasycznej baterii testów DIEHARD, dostępne jako narzędzie open-source do badania generatorów losowości. Zostało ono wskazane jako dobry wybór na „drugi poziom” walidacji po testach własnych lub testach opartych na NIST. 

W kontekście projektu Dieharder może zostać wykorzystany do niezależnego sprawdzenia generatorów po przejściu testów podstawowych. Dzięki temu możliwe będzie zestawienie wyników uzyskanych w notebooku z wynikami pochodzącymi z zewnętrznego, uznanego narzędzia testowego.

### 5. Wizualizacja i raportowanie wyników

Oprócz testów statystycznych istotnym elementem procesu weryfikacji będzie wizualizacja wyników. Zgodnie z założeniami zawartymi w szkicu projektu moduł weryfikacji powinien obejmować nie tylko same testy, ale również prezentację rozkładu danych oraz raport wynikowy zawierający wartości statystyk, p-value oraz końcową ocenę typu pass/fail. 

W praktyce oznacza to, że dla każdego generatora planowane jest przygotowanie:
- tabeli z podstawowymi statystykami,
- wykresów rozkładu,
- zestawienia wyników testów,
- krótkiej interpretacji uzyskanych rezultatów.

Takie podejście pozwoli nie tylko stwierdzić, czy generator przechodzi dany test, ale również lepiej zrozumieć charakter ewentualnych odchyleń i różnice pomiędzy poszczególnymi klasami generatorów.
